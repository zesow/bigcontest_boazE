# 장르 더미변수 처리

## 장르의 집합을 만들어준다

```python
genre_iter = (set(x.split(',')) for x in rawData.genre)
genres = sorted(set.union(*genre_iter))
```

## (원 데이터 x 장르 갯수) 크기의 matrix 만들어준다

```python
dummies = pd.DataFrame(np.zeros((len(rawData),len(genres))),columns=genres)
```

### raw loop 돌면서 있는 거만 1로 만들어준다.

```python
for i, gen in enumerate(rawData.genre):
    dummies.ix[i,gen.split(',')] = 1
```

## 더미변수 처리한 걸 옆에 붙여준다

```python
processed_dummy = rawData.join(dummies.add_prefix(''))
```

```python
processed_dummy
```

# 관람가 더미변수 처리

```python
age_iter = (set(x) for x in processed_dummy.grade)
ages = set(processed_dummy.grade)
```

```python
dummies = pd.DataFrame(np.zeros((len(processed_dummy),len(ages))),columns=ages)
```

```python
for i, gen in enumerate(processed_dummy.grade):
    dummies.ix[i,gen] = 1
```

```python
processed_dummy_age = processed_dummy.join(dummies.add_prefix(''))
```
